```python
def split_litho(s):
    if s is None:
        return []
    return [x.strip().lower() for x in str(s).split(";")]
    
geol_nt["litho_list"] = geol_nt["LITHOLOGY"].apply(split_litho)
```

```python
def litho_to_group(s):
    if "regolith" in s:
        return "regolith"
    if "sedimentary siliciclastic" in s:
        return "sed_siliciclastic"
    if "sedimentary carbonate" in s:
        return "sed_carbonate"
    if "felsic intrusive" in s:
        return "igneous_felsic_intrusive"
    if "mafic volcanic" in s:
        return "igneous_mafic_volcanic"
    if "mafic intrusive" in s:
        return "igneous_mafic_intrusive"
    if "felsic volcanic" in s:
        return "igneous_felsic_volcanic"
    if "metasedimentary" in s:
        return "metasedimentary"
    if "metamorphic" in s:
        return "metamorphic"
    return None

geol_nt["litho_groups"] = geol_nt["litho_list"].apply(
    lambda xs: sorted(set(filter(None, (litho_to_group(x) for x in xs))))
)
```

```python
all_groups = sorted({g for xs in geol_nt["litho_groups"] for g in xs})

for g in all_groups:
    geol_nt[f"litho_{g}"] = geol_nt["litho_groups"].apply(lambda xs: int(g in xs))
```

```python
def age_eon(text):
    if not text:
        return "Unknown"

    t = text.lower()

    # --- ARCHEAN ---
    if any(k in t for k in [
        "archean", "archaean", "neoarchean"
    ]):
        return "Archean"

    # --- PROTEROZOIC ---
    if any(k in t for k in [
        "proterozoic",
        "paleoproterozoic", "mesoproterozoic", "neoproterozoic",
        "siderian", "rhyacian", "orosirian", "statherian",
        "calymmian", "ectasian", "stenian",
        "cryogenian", "tonian", "ediacaran"
    ]):
        return "Proterozoic"

    # --- PHANEROZOIC ---
    if any(k in t for k in [
        # Periods
        "cambrian", "ordovician", "silurian", "devonian",
        "carboniferous", "permian",
        "triassic", "jurassic", "cretaceous",
        "paleogene", "neogene",
        "cenozoic", "quaternary", "holocene",

        # Epochs
        "miocene", "pliocene", "eocene", "paleocene",

        # Australian / international stages commonly in data
        "boomerangian", "mindyallan", "datsonian",
        "furongian", "tremadocian", "arenig",
        "darriwilian", "emsian", "frasnian",
        "tournaisian", "visean", "namurian",
        "gzhelian", "sakmarian", "artinskian",
        "anisian", "tithonian", "valanginian",
        "albian", "aptian", "cenomanian",
        "maastrichtian", "barremian",
        "ordian", "templetonian", "idamean",
        "llanvirn", "undillan", "westphalian",
        "stephanian"
    ]):
        return "Phanerozoic"

    return "Unknown"
```

```python
def age_period(text):
    if not text:
        return "Unknown"

    t = text.lower()

    # --- Phanerozoic: Periods & common stages ---
    phanerozoic = [
        # Cenozoic
        "holocene", "quaternary", "miocene", "pliocene",
        "eocene", "paleocene", "cenozoic",

        # Mesozoic
        "cretaceous", "jurassic", "triassic",
        "albian", "aptian", "cenomanian", "maastrichtian",
        "valanginian", "tithonian", "barremian",

        # Paleozoic
        "permian", "carboniferous", "devonian",
        "silurian", "ordovician", "cambrian",
        "tournaisian", "visean", "namurian",
        "westphalian", "stephanian",
        "frasnian", "emsian",
        "furongian", "tremadocian", "arenig",
        "darriwilian", "llanvirn",
        "ordian", "templetonian", "idamean",
        "undillan", "boomerangian", "mindyallan", "datsonian"
    ]

    for p in phanerozoic:
        if p in t:
            return p.capitalize()

    # --- Proterozoic subdivisions ---
    proterozoic = [
        "ediacaran", "cryogenian", "tonian",
        "stenian", "ectasian", "calymmian",
        "statherian", "orosirian", "rhyacian", "siderian",
        "neoproterozoic", "mesoproterozoic", "paleoproterozoic"
    ]

    for p in proterozoic:
        if p in t:
            return p.capitalize()

    # --- Archean ---
    if "archean" in t or "archaean" in t:
        return "Archean"

    return "Unknown"
```

## Create centroid + buffer 5 km for each polygon geology

```python
from shapely.ops import unary_union

fault_union = unary_union(faults_nt.geometry)
contact_union = unary_union(contacts_nt.geometry)

geol_nt["dist_to_fault_m"] = geol_nt.geometry.centroid.distance(fault_union)
geol_nt["dist_to_contact_m"] = geol_nt.geometry.centroid.distance(contact_union)
```

```python
import geopandas as gpd

R = 5000  # 5 km

# 1) centroids + buffers
centroids = geol_nt.geometry.centroid
buffers = gpd.GeoDataFrame(geometry=centroids.buffer(R), crs=geol_nt.crs)


# 2) FAULT length within 5km

fault_join = gpd.sjoin(
    faults_nt[["geometry"]],
    buffers,
    predicate="intersects"
)

# calculate length before groupby
fault_join["seg_len"] = fault_join.geometry.length  # metres (CRS projected)
fault_len = fault_join.groupby("index_right")["seg_len"].sum()

geol_nt["fault_len_5km"] = geol_nt.index.map(fault_len).fillna(0.0)

# 3) CONTACT length within 5km

contact_join = gpd.sjoin(
    contacts_nt[["geometry"]],
    buffers,
    predicate="intersects"
)

contact_join["seg_len"] = contact_join.geometry.length
contact_len = contact_join.groupby("index_right")["seg_len"].sum()

geol_nt["contact_len_5km"] = geol_nt.index.map(contact_len).fillna(0.0)
```

### Calculate function azimuth/strike for LineString/MultiLineString

```python
import numpy as np
import pandas as pd
from shapely.geometry import LineString, MultiLineString

def azimuth_0_180(geom):
    """Return strike azimuth in [0,180) from a LineString/MultiLineString."""
    if geom is None or geom.is_empty:
        return np.nan

    
    if isinstance(geom, MultiLineString):
        geom = max(list(geom.geoms), key=lambda g: g.length)

    if not isinstance(geom, LineString):
        return np.nan

    coords = list(geom.coords)
    if len(coords) < 2:
        return np.nan

    x1, y1 = coords[0]
    x2, y2 = coords[-1]

    
    dx = x2 - x1
    dy = y2 - y1
    ang = np.degrees(np.arctan2(dx, dy))  # North-based
    az = ang % 180
    return az
```

```python
faults_nt = faults_nt[faults_nt.is_valid].copy()
faults_nt["azimuth"] = faults_nt.geometry.apply(azimuth_0_180)
```

```python
faults_nt["azimuth"]
```

```python
bins = [0, 22.5, 67.5, 112.5, 157.5, 180]
labels = ["N-S_0", "NE-SW", "E-W", "NW-SE", "N-S_180"]  # wrap

faults_nt["az_bin"] = pd.cut(
    faults_nt["azimuth"],
    bins=bins,
    labels=labels,
    include_lowest=True
)
```

```python
faults_nt["az_bin"]
```

```python
import geopandas as gpd

R = 5000

geol_nt = geol_nt[geol_nt.is_valid].copy()

centroids = geol_nt.geometry.centroid
buffers = gpd.GeoDataFrame(geometry=centroids.buffer(R), crs=geol_nt.crs)

join = gpd.sjoin(
    faults_nt[["geometry", "az_bin"]],
    buffers,
    predicate="intersects"
)

join["seg_len"] = join.geometry.length

orient_len = (
    join.groupby(["index_right", "az_bin"])["seg_len"]
        .sum()
        .unstack(fill_value=0.0)
)


orient_len = orient_len.rename(columns={
    "N-S": "fault_len_NS_5km",
    "NE-SW": "fault_len_NE_SW_5km",
    "E-W": "fault_len_EW_5km",
    "NW-SE": "fault_len_NW_SE_5km",
})

geol_nt = geol_nt.join(orient_len, how="left").fillna(0.0)
```
